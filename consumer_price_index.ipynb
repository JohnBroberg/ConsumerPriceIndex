```python
#Consumer Price Index from Bureau of Labor Statistics
#created by John Broberg on 2022-05-22

#cd
#cd Desktop/ConsumerPriceIndex
#git pull https://github.com/JohnBroberg/ConsumerPriceIndex.git


import pandas as pd
import numpy as np
import seaborn as sns

url = 'http://www.bls.gov/news.release/cpi.t01.htm'
dfs = pd.read_html(url, header=0)

print(len(dfs))
```

```python
type(dfs[0])
```

```python
df = dfs[0]
df
```

```python
df1 = df[['Expenditure category', 'Unadjusted percent change']]
header1 = df1.iloc[0,0]
header2 = df1.iloc[0,1]
df1.columns = [header1, header2]

df1 = df1.dropna() \
    .drop(index=[0,40]) \
    .astype({df1.columns[1]: 'float'}) \
    .sort_values(df1.columns[1], ascending=False) \
    .style.background_gradient(axis=0, cmap='Reds') \


print('\nConsumer Price Index for All Urban Consumers (CPI-U): \
      \nU.S. city average, by expenditure category, \
      \nin order of Unadjusted percent change')
df1
```
