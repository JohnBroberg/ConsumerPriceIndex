```python
import pandas as pd
import numpy as np

url = 'http://www.bls.gov/news.release/cpi.t01.htm'
dfs = pd.read_html(url, header=0)

print(len(dfs))
```

```python
type(dfs[0])
```

```python
df = dfs[0]
df
```

```python
df1 = df[['Expenditure category', 'Unadjusted percent change']]
df1 = df1.rename(columns={'Unadjusted percent change':'Unadjusted percent change Apr.2021-Apr.2022'})
df1 = df1.dropna()
df1 = df1.drop(index=[0,40])
df1 = df1.astype({'Unadjusted percent change Apr.2021-Apr.2022': 'float'})
df1 = df1.sort_values('Unadjusted percent change Apr.2021-Apr.2022', ascending=False)
df1 = df1.style.background_gradient(axis=0, cmap='Reds')

df1
```
